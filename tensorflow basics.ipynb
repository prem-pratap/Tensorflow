```python
import tensorflow as tf
```

```python
#buliding computational graph
n1=tf.constant(3.5,tf.float32)
n2=tf.constant(4.0,tf.float32)
n3=n1*n2
print(n1,n2)
```

```python
#running session
sess=tf.Session()
output=sess.run([n1,n2])
o2=sess.run(n3)
#for graph visualization we need to create a class object to write summaries, one such class is FileWriter("directory name")
File_writer=tf.summary.FileWriter("/home/prempratap/Desktop/ML/tensorflow/graph1",sess.graph)
print(output)
print(o2)
sess.close()

#we can use with to run session and this will run only the block of code
with tf.Session() as sess:
    print(sess.run([n1,n2]))
```

```python
#placeholder are variables used for external inputs.A graph can be parameterized to accept external inputs known as placeholder
#A placeholder is promise to provide a value later.
import tensorflow as tf
node1=tf.placeholder(tf.float32)
node2=tf.placeholder(tf.float32)
adder=node1+node2
sess=tf.Session()
print(sess.run(adder,{node1:[1,2,3],node2:[4,5,6]}))
File_writer=tf.summary.FileWriter("/home/prempratap/Desktop/ML/tensorflow/graph2",sess.graph)
sess.close()
```

```python
#variable are used to hohld update parameters. These are required to modify the graph for the new output for the same input
import tensorflow as tf
#model parameters
w=tf.Variable([.3],tf.float32)
b=tf.Variable([-.3],tf.float32)
#inputs and outputs
x=tf.placeholder(tf.float32)
linear_model=w * x +b
#y hold actual output which we already know
y=tf.placeholder(tf.float32)
#loss
squared_delta=tf.square(linear_model-y)
loss=tf.reduce_sum(squared_delta)
#optimize our model
optimizer=tf.train.GradientDescentOptimizer(0.01)#0.01 is learning rate or steps by which we change our variable
train=optimizer.minimize(loss)
init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
for i in range(1000):
        print(sess.run(train,{x:[1,2,3,4],y:[0,-1,-2,-3]}))
print(sess.run([w,b]))
#print(sess.run(loss,{x:[1,2,3,4],y:[0,-1,-2,-3]}))
File_writer=tf.summary.FileWriter("/home/prempratap/Desktop/ML/tensorflow/graph3",sess.graph)
sess.close()
```
